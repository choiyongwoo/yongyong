<h1>용어</h1>
그래프는 여러 노드들이 간선으로 연결된 추상 네트워크를 뜻함<br>
즉, 그래프는 노드와 간선의 집합으로 정의되며 이를 수식으로 쓰면 다음과 같음<br>
G= (V, E)<br><br>

여기서 노드 집합 V는 임의의 유한 집합이고, 간선 집합 E는 노드 쌍들의 집합<br>
예를 들어 노드 집합 V= [a, b, c, d]이고 간선집합 E={{a, b}, {b, c}, {c, d}, {d, a}}인 그래프는 다음과 같음

![](IMG/graph_basics1.jpg)

<h3>그래프 방향</h3>
그래프에는 방향이 있는 유향 그래프와 방향이 없는 무향 그래프가 있음<br><br>

무향 그래프는 말 그대로 간선에 방향이 지정되어 있지 않음<br>
간선으로 연결된 노드는 서로 인접해 있으며, 이웃이라고 함

![](IMG/graph_basics2.jpg)

유향 그래프의 경우 순서가 존재하므로 말단 노드가 존재<br>
노드 u와 v사이의 간선은 u에서 v까지의 간선 (u, v)또는 v에서 u까지의 간선 (v, u)가 있다<br>
유향 그래프 G에서 함수 E(G)는 V(G)에 대한 관계라고도 할 수 있음

![](IMG/graph_basics3.jpg)

<h3>부분 그래프</h3>
부분 그래프는 그래프 G에서 집합 V와 E로 구성된 그래프의 일부<br>
다음과 같은 그래프 G가 있다고 하자

![](IMG/graph_basics4.jpg)

다음 예들이 부분 G의 부분그래프

![](IMG/graph_basics5.jpg)

위의 부분 그래프를 차례로 G1, G2, G3이라고 했을 때, G3은 G2의 유도 부분 그래프라고 함<br><br>

한편, 신장 부분 그래프는 원본 그래프의 모든 노드를 포함하는 부분 그래프를 말함<br>
다음 그림에서 간선을 굵게 표시한 부분이 신장 부분 그래프

![](IMG/graph_basics6.jpg)

<h3>완전 그래프</h3>
완전 그래프는 그래프의 모든 노드가 서로 인접한 그래프를 말함<br>

![](IMG/graph_basics7.jpg)

<h3>차수</h3>
한 노드에 이어져 있는 간선의 수를 차수라고함<br>
차수가 0인 노드는 고립되었다고 부름<br>
유향 그래프의 경우는 입력 차수와 출력 차수로 나눌 수 있음<br>
입력 차수는 한 노드로 들어오는 간선 수를 말하며, 출력 차수는 한 노드에서 나가는 간선 수를 말함

<h3>경로, 보행, 순환</h3>
그래프에서 경로는 간선이 어느 노드도 다시 방문하지 않고, 노드가 일렬로 연결된 부분 그래프, 유향 그래프에서 경로는 간선의 방향을 따름<br><br>

보행은 노드와 간선을 번갈아가며 반복적으로 방문<br>
경로는 노드와 간선이 모두 중복되지 않은 보행과 같음<br><br>

순환은 경로와 같지만 마지막에 연결된 간선의 노드가 다시 첫 번째 노드에 연결

<h3>경로 길이</h3>
경로 또는 보행의 길이는 간선의 수와 동일

<h3>가중 그래프</h3>
가중 그래프는 간선에 가중치가 있는 그래프<br>
경로 또는 순환의 가중치는 해당하는 간선들의 가중치의 총합<br>
가중 그래프가 아닌 경우에는 경로와 순환의 가중치가 간선의 수와 같음

<h4>평면 그래프</h4>
평면 그래프는 간선을 서로 횡단하지 않고 평면에 그릴 수 있는 그래프<br>
이 그래프는 간선에 의해 경계면을 가짐<br>
연결된 평면 그래프의 오일러 공식에 따르면 V-E+F=2(V:노드수, E:간선수, F:면수)

![](IMG/graph_basics8.jpg)

두 간선이 교차하기 때문에 평면 그래프가 아닌 것 같지만, 위 그래프는 평면 그래프<br>
다음 그림과 같이 변환을 통해 다양한 형태로 평면에 다시 나타낼 수 있기 때문

![](IMG/graph_basics9.jpg)

<h4>순회</h4>
순회는 그래프에 연결된 모든 요소를 탐색하는 일<br>
순회에서 중요한 것은 (아직 방문하지 않은)노드의 순회 순서

<h4>강한 연결 요소</h4>
무향 그래프는 모든 노드에서 다른 모든 노드로 가는 경로가 존재할 때 연결 되어 있다고 함<br>
유향 그래프도 마찬가지<br><br>

연결 요소는 모든 노드가 연결된 최대 부분 그래프를 말함<br>
연결 요소는 깊이 우선 탐색과 너비 우선 탐색 같은 순회 알고리즘을 사용하여 찾을 수 있음<br><br>

유향 그래프에서 모든 노드에서 다른 모든 노드까지의 경로가 있으면, 강하게 연결되었다고 함<br>
이때 강한 연결 요소는 강하게 연결된 최대 하위 그래프를 말함<br>

<h4>트리와 포레스트</h4>
포레스트는 순환이 없는 그래프<br>
트리는 비순환적이고 연결되어 있는 유향 그래프<br>
포레스트는 하나 이상의 트리로 구성되어 있음, 즉 서로 독립적인 트리의 모임<br><br>

트리에서 두 노드는 정확히 하나의 경로로 연결됨<br>
트리에 새로운 간선을 하나 추가하면 순환이 생기고, 어떤 간선을 제거하면 연결되지 않은 요소가 생성
