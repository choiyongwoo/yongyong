<h2>복잡도 종류</h2>
복잡도 종류는 계산에 드는 복잡도에 따라 문제들을 분류해놓은 집합<br>
이때 필요한 것은 복잡도 환산(변환)으로서 한 문제를 원래 문제만큼 어렵게 다른 문제로 변환하는 것을 뜻함<br>
가장 일반적으로 사용되는 환산은 다항 시간 환산이며, 이는 환산 절차가 다항시간이 걸린다는 것을 의미<br>

<h3>P</h3>
복잡도에서 P는 결정론적 튜링 기계로 다항 시간안에 풀 수 있는 판정 문제를 모아놓은 복잡도 종류<br>
결정론적이란말에 주목해 쉽게 풀어 말하자면 계산의 각 단계에서 단 한 가지 가능성만을 고려할 수 있는 알고리즘으로 다항 시간 안에 풀 수 있는 문제<br>
어떤 문제를 결정론적 문제로 바꿀 수 있다면, 그 결과는 P에 속함<br>

<h3>NP</h3>
NP는 비결정론적 튜링 기계로 다항 시간 안에 풀 수 있는 판정 문제의 집합<br>
P와 구분되는 것은 비결정론적이라는 점으로, 각 계산단계에서 여러 가능성을 동시에 고려할 수 있는 알고리즘을 통해 다항 시간 안에 풀 수 있다는 뜻<br><br>
어떤 문제가 적어도 NP에 속하는 어떤 문제만큼 어렵다면(다항 시간 환산으로 판정될 수 다면)NP-난해 라고함<br>
다른 정의로는, 모든 NP문제가 다항 시간에 어떤 문제로 환산될 수 있다면 그 문제를 NP-난해 라고 함<br>
예로, 그래프를 통해 최단 경로를 찾는 외판원 문제는 NP-난해에 속함<br><br><br>
NP-난해 문제 자체는 NP문제가 아닐 수도 있음<br>
NP문제이자 NP-난해인 문제를 NP-완전이라고 함<br>
NP-완전은 NP집합에 속하는 판정 문제 중에서 가장 어려운 문제의 부분 집합으로 모든 NP문제는 다항 시간 내에 NP-완전 문제로 환산할 수 있음
<br><br>
NP에 속하는 문제는 그 문제의 대답이 '예'라는 근거가 주어졌을 때 그것이 옳은 근거임을 다항식 시간에 확인해줄 수 있는 문제<br>
반면,P에 속하는 문제는 그 문제의 질문이 주어졌을 때, Yes 또는 No의 대답을 다항식 시간에 해줄 수 있는 문제<br><br>
NP 문제는 대답이 Yes라는 근거가 제시되었을 때 그것이 맞는다는 것을 다항식 시간에 확인만 하면 될 뿐 대답이 No일 경우는 어떻게 하든 상관 없음<br><br>
P문제는 질문 자체에 대해서 다항식 시간에 Yes 또는 No대답을 해야함<br>
질문에 대답을 한다는 것은 '문제를 푼다'라는 것과 일치함

<h3>P=NP?</h3>
P-NP문제는 복잡도 종류 P와 NP가 같은지 여부에 대한 컴퓨터 과학 분야의 미해결 문제<br>
앞에서 언급했듯이 P는 결정론적 튜링 기계로, NP는 비결정론적 튜링기계로 다항 시간에 답을 구할 수 있는 문제의 집합<br>
결정론적 튜링 기계에 사용한 프로그램은 비결정론적 튜링 기계에도 적용할 수 있으므로 P는 NP의 부분 집합이 된다<br>
하지만 P와 NP가 같은 집합인지, 아니면 P가 NP의 진부분집합인지 여부는 아직 밝혀지지 않음<br><br>

NP-완전 문제 중 하나라도 P에 속한다는 것을 증명한다면 모든 NP문제가 P에 속하기 때문에, P-NP문제가 P=NP가 되고 반대로 NP-완전 문제 중 하나가 P에 속하지 않는다는 것이 증명된다면 P=NP에 대한 반례가 되어 P!=NP가 됨

co-NP는 NP문제와 보완 관계인 문제 종류<br>
문제 x가 co-NP에 속한다는 것은 그 보완 문제인 'x바' 가 NP에 속한다는것과 동치 즉, co-NP는 '예'의 반대인 '아니요'경우(반례)에 대해 효율적으로 검증할 수 있는 증명이 있는 문제의 집합<br>
NP-완전 문제 중에 부분집합 합 문제가 있음<br>
이 문제의 보완 문제는 정수 유한 집합이 주어질 때, 공집합이 아닌 모든 부분집합은 원소를 다 더했을 때 0이 아닌지를 구하는 문제가 됨<br>
'아니요'의 경우에 대한 증명을 하려면, 합이 0이 되고 공집합이 아닌 부분집합을 찾으면 되므로 증명을 검증하기가 쉬워짐<br><br>

다항 시간에 풀 수 있는 문제인 P는 NP와 co-NP 모두의 부분집합<br>
