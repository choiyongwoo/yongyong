<h2>함수의 실행 시간</h2>
알고리즘의 실행 시간(런타임) 분석<br>
알고리즘에 재귀 호출이 없다면, 자료구조와 코드 블록 흐름만 파악하면 됨(하나씩 실행되는 코드 블록의 시간복잡도를 계산하면 됨)<br>
중첩 반복문의 경우 시간복잡도는 배가 됨<br><br>

알고리즘에 재귀 호출이 있다면 재귀식을 사용하여 실행 시간을 찾을 수 있음<br>
함수에 대한 재귀식을 작성할 때는 베이스 케이스(O(1)이어야 하므로 T(1)=1)와 일반 케이스 두 방정식을 작성해야함<br>
피보나치 수열에서 n번째 요소를 찾는 알고리즘 예제

<h2>알고리즘 성능 평가</h2>
알고리즘 성능을 평가하기 위해 '복잡도'라는 척도 사용<br>
그중 시간 복잡도와 공간 복잡도의 개념이 나오며, 동일한 기능을 수행하는 알고리즘이 있을 때, 복잡도가 낮을 수록 좋은 알고리즘<br>
시간 복잡도: 특정한 크기의 입력에 대하여 알고리즘의 수행시간 분석<br>
공간 복잡도: 특정한 크기의 입력에 대하여 알고리즘의 메모리 사용량 분석

<h3>시간 복잡도</h3>
시간 복잡도는 특정 알고리즘이 어떤 문제를 해결하는데 걸리는 시간을 의미<br>
같은 결과를 갖는 프로그래밍 소스도 작성 방법에 따라 걸리는 시간이 달라지며, 같은 결과를 갖는 소스라면 시간이 적게 걸리는 것이 좋은 소스<br><br>
<h4>빅-오 표기법</h4>
시간 복잡도에는 빅-오 표기법이라는 개념이 있음<br>
예를 들어, 동전을 튕겨 뒷면이 나올 확률을 이야기 할 때 운이 좋으면 1번에 뒷면이 나오지만 운이 안 좋다면 n번 만큼 동전을 튕겨야 하는 경우가 발생, 이 최악의 경우를 계싼하는 방식을 빅-오(big-o)표기법이라 부름<br>

![](IMG/asymptotic_analysis7.png)

O(1): 입력 데이터의 크기에 상관없이 언제나 일정한 시간이 걸리는 알고리즘, 데이터가 얼마나 증가하든 성능에 영향을 거의 미치지 않음<br><br>
O(log₂ n): 입력 데이터의 크기가 커질수록 처리 시간이 로그 만큼 짧아지는 알고리즘, 예를들어 데이터가 10배가 되면, 처리 시간은 2배, 이진 탐색이 대표적이며, 재귀가 순기능으로 이루어지는 경우도 해당<br><br>
O(n): 입력 데이터의 크기에 비례해 처리 시간이 증가하는 알고리즘, 예를 들어 데이터가 10배가 되면 처리 시간도 10배, 1차원 for문<br><br>
O(n log₂ n): 데이터가 많아질수록 처리시간이 로그(log)배만큼 더 늘어나는 알고리즘, 예를 들어 데이터가 10배가 되면, 처리 시간은 약 20배, 정렬 알고리즘 중 병합 정렬, 퀵 정렬이 대표적<br><br>

O(n²): 데이터가 많아질수록 처리시간이 급수적으로 늘어나는 알고리즘, 예를 들어 데이터가 10배가 되면, 처리시간은 최대 100배, 이중 루프가 대표적이며, 단 m이 n보다 작을때는 반드시 O(nm)으로 표시하는 것이 바람직<br><br>

O(2ⁿ): 데이터량이 많아질수록 처리시간이 기하급수적으로 늘어나는 알고리즘<br>
대표적으로 피보나치 수열이 있으며, 재귀가 역기능을 할 경우도 해당

<h3>공간 복잡도</h3>
공간 복잡도란 작성한 프로그램이 얼마나 많은 공간을 차지하느냐를 분석하는 방법<br>
최근에 대용량 시스템이 보편화되면서, 공간 복잡도보다는 시간 복잡도가 우선

<h4>공간 복잡도 줄이기</h4>
공간 복잡도를 결정하는 것은 보통 배열의 크기가 몇인지, 얼마 만큼의 동적 할당인지, 몇번의 호출을 하는 재귀 함인지 등이 공간 복잡도에 영향을 끼침<br>
프로그램에 필요한 공간은 크게<br>
1. 고정 공간<br>
2. 가변 공간<br>
이 있는데, 시간적인 측면을 무시하고 공간 복잡도만 고려한다면 고정 공간보다는 가변 공간을 사용할 수 있는 자료 구조가 더 효율적<br><br>
함수 호출시 할당되는 지역변수들이나 동적 할당되는 객체들도 모두 공간이 필요<br>
특히 재귀함수의 경우 매 함수 호출마다 함수의 매개변수, 지역변수, 함수의 복귀 주소를 저장할 공간이 필요해서 재귀적으로 짤 수 있고, 반복문으로도 짤 수 있는 경우에는 반복문으로 짜는 것이 더 효율적

```python
def find_fibonacci_seq_rec(n):
    if n<2:
        return n
    return find_fibonacci_seq_rec(n-1)+find_fibonacci_seq_rec(n-2)

if __name__== '__main__':
    print(find_fibonacci_seq_rec(5))
    
```
