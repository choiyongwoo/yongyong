```python
#tuple
#불변한 순서가 있는 객체의 집합
#list형과 비슷하지만 한번 생성되면 값을 변경할 수 없음
t= (1, 'korea', 3.5, 1)
print(t)
print(type(t))

#순서가 있기때문에 인덱스로 접근 가능
print(t[0])
print(t[1])

#len 내장함수로 길이 구할 수 있음
print(len(t),'\n')

#list와 마찬가지로 for loop를 돌 수 있음
for i in t:
    print(i)
    
#+연산으로 tuple을 추가할 수 있음
t= t+(3, 5)
print('\n',t)

#*연산으로 tuple을 반복할 수 있음

#tuple속에 tuple이 포함될 수 있음
a= ((1, 2), (3, 4), (5, 9))
print(a[2])
print(a[2][1])

#tuple이 하나의 원소만 존재하는 경우 tuple이 되지 않음
#한개의 원소 뒤에 콤마를 찍어주면 tuple이 유지
h= (350)
print('\n', type(h))
h=(350, )
print(type(h))

#괄호가 없어도 tuple 입력 가능
p= 1, 3, 2, 5, 7
print(type(p))

#tuple을 이용하여 함수에서 여러 값을 한꺼번에 리턴 가능
def minmax(items):
    return min(items), max(items)
minmax([7, 5, 2, 1, 11, 15, 55])

#tuple을 이용하여 변수를 한꺼번에 할당할 수 있음
lower, upper= minmax([7, 5, 2, 1, 11, 15 ,55]) #(lower, upper)와 동일
print('\n', lower)
print(upper)

#tuple속에 tuple을 이용해 할당가능
(a, (b,(c, d)))= (4, (3,(2, 1)))
print(a, b, c, d)

#tuple의 해체 할당기능을 통해 두 변수의 값을 바꿔줄 수 있음
a= '감자'
b= '고구마'
a, b= b, a
print(a, b)

#tuple변환 - tuple(객체)로 tuple로 변형할 수 있음
print(tuple([1, 7, 5, 3, 9]))
print(tuple('abcde'))

#in절로 포함되어 있는지 체크
print()'a' in ('a', 'b', 'c')
```
